2. Вектор размером 11 со значениями от 0 до 5. Расчитываем на его основе вектор y(y=x^2)

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

3. Построение графиков

3.1 Функциональный метод

```python


plt.plot(x,y) #build plot
plt.xlabel('X Label') # add name of x axis
plt.ylabel('Y Label') # add name of y axis
plt.title('Title') # add title

```

Построение графиков рядом

```python
# create multiply plots

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

3.2 Объектно-ориентированный метод

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

Одновременное построение нескольких графиков, добавляем их к объекту с указанием области видимости

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(x,y,'b')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title')
```

Функция subplots при создании графика, как объекта

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

Указываем количество строк и столбцов при создании объекта subplots()

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

Случай наложения графиков

```python
fig, axes = plt.subplots(nrows=4,ncols=4)
```

```python
Функция для избежания наложения графиков
```

```python
fig, axes = plt.subplots(nrows=4,ncols=4)
plt.tight_layout()
```

Задавание легенды графику

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2,label="x**2")
ax.plot(x, x**3,label="x**3")
ax.legend()
```

Изменение размера графика и DPI 

```python
fig = plt.figure()
fig = plt.figure(figsize=(8,4),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2,label="x**2")
ax.plot(x, x**3,label="x**3")
ax.legend()
```

Изменение цвета графика

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,color = 'b')
ax.plot(x,x**3,color = 'g')
```

Изменение типа и ширины линии

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,lw=2,ls = '--')
```

Изменение типа, размера и цвета маркеров

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,marker = 'o', markersize = 10, markerfacecolor = 'y')
```
